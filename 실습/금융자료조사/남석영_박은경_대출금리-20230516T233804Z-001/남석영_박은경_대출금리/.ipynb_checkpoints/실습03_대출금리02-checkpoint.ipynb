# 예금 & 대출 금리

```python
예금이자 : 은행이 예금하는 고객에게 지급하는 이자 (기준금리 + 가산금리 + 우대금리) 
대출이자 : 은행이 대출하는 고객에게 요구하는 이자 (기준금리 + 가산금리 - 우대금리)

가산금리(Spread Rate) : 기준금리에 신용도 등의 조건에 따라 덧붙이는 금리
우대금리(Prime Rate) : 우량기업(개인)에게 적용하는 금리

# 무위험이자률 (MBS,단기채권)기준 + 0.01 a = 2.52 -> 예대마진 
```

```python
rate=pd.read_csv("./datasets/예금대출.csv")
rate=rate.rename(columns={"변환":"날짜","원자료":"예금_금리","원자료.1":"대출_금리"})
# rate=rate.set_index('날짜')
rate.날짜=pd.to_datetime(rate.날짜) 
rate.head()
```

<img src = "./datasets/예금대출.png">

```python
rate['금리차이']=rate.대출_금리 - rate.예금_금리
rate.head()
```

## 예대마진 : 금융기관의 수익성 지표 

```python
https://terms.naver.com/entry.naver?docId=1199689&cid=40942&categoryId=31826

금융기관의 수입 = 대출이자 - 예금이자
```

```python
rate.금리차이.mean()  
```

```python
fig = plt.figure(figsize=(8,4))
ax = fig.add_subplot()
ax.patch.set_facecolor('white')

x = rate.날짜
y_1 = rate['예금_금리']
y_2 = rate['대출_금리']

plt.plot(x, y_1, label='예금_금리', color='purple', linestyle=':',alpha=0.7)
plt.plot(x, y_2, label='대출_금리', color='c', linestyle='-.',alpha=0.7 )

plt.title('[ 기준금리 : 예금 / 대출 ]', fontsize= 10)   
plt.xlabel('날짜', fontsize=10)  
plt.ylabel('금리', fontsize=10) 
 
plt.legend(['예금 금리', '대출 금리'], fontsize=10)  
plt.rcParams["legend.facecolor"] = 'white'

plt.grid(False, axis='x', color='gray', alpha=0.2)
plt.show()
```

## 콜금리

```python
call=pd.read_csv("./datasets/콜금리.csv")
call=call.rename(columns={"변환":"날짜","원자료": "기준금리","원자료.1":"무담보 콜 금리","원자료.2":"담보 콜 금리"})
call.날짜=pd.to_datetime(call.날짜)
call.head(3)
```

```python
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot()
ax.patch.set_facecolor('white')

x = call.날짜
y_1 = call['기준금리']
y_2 = call['무담보 콜 금리']
y_3 = call['담보 콜 금리']

plt.plot(x, y_1, label='기준금리', color='purple', alpha=0.7)
plt.plot(x, y_2, label='무담보 콜 금리', color='c',alpha=0.7 )
plt.plot(x, y_3, label='담보 콜 금리', color='g', alpha=0.7 )

plt.title('[ 기준금리 : 무담보 / 담보 콜 금리 ]', fontsize= 10)   
plt.xlabel('날짜', fontsize=10)  
plt.ylabel('금리', fontsize=10) 
 
plt.legend(['기준금리','무담보 콜 금리', '담보 콜 금리'], fontsize=10)  
plt.rcParams["legend.facecolor"] = 'white'

plt.grid(False, axis='x', color='gray', alpha=0.2)
plt.show()
```

```python
[담보콜금리 이율이 높은 이유]
금융기관의 신용도가 낮은 경우 결제 안정성이 낮아 담보로 유가 증권 등을 맡기고 거래

대부분의 콜금리는 무담보로 이루어짐
```

## KORIBOR

```python
koribor = pd.read_csv("./datasets/코리보.csv")
koribor=koribor.rename(columns={"변환":"날짜","원자료": "기준금리","원자료.1":"KORIBOR 3개월","원자료.2":"KORIBOR 6개월","원자료.3":"KORIBOR 12개월"})
koribor.날짜=pd.to_datetime(koribor.날짜)
koribor.head(3)
```

```python
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot()
ax.patch.set_facecolor('white')

x = koribor.날짜
y_1 = koribor['기준금리']
y_2 = koribor['KORIBOR 3개월']
y_3 = koribor['KORIBOR 6개월']
y_4 = koribor['KORIBOR 12개월']

plt.plot(x, y_1, label='기준금리', color='purple', alpha=0.7)
plt.plot(x, y_2, label='KORIBOR 3개월', color='c',alpha=0.7 )
plt.plot(x, y_3, label='KORIBOR 6개월', color='g', alpha=0.7 )
plt.plot(x, y_4, label='KORIBOR 12개월', color='b', alpha=0.7 )

plt.title('[ 기준금리 : KORIBOR 3, 6, 12개월 금리 ]', fontsize= 10)   
plt.xlabel('날짜', fontsize=10)  
plt.ylabel('금리', fontsize=10) 
 
plt.legend(['기준금리','KORIBOR 3개월', 'KORIBOR 6개월','KORIBOR 12개월'], fontsize=10)  
plt.rcParams["legend.facecolor"] = 'white'

plt.grid(False, axis='x', color='gray', alpha=0.2)
plt.show()
```
