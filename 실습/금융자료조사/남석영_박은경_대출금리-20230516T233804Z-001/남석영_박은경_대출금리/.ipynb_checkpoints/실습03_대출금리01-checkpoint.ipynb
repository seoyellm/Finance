# DataLoad

```python
loan_com = pd.read_csv("./대출금리/예금은행 대출금리_기업.csv")
loan_com = pd.DataFrame(loan_com)
loan_com = loan_com.drop(columns = ["통계표", "단위", "변환"], axis=1)
loan_com = loan_com.set_index("계정항목")
loan_com =loan_com.T
loan_com.head(1)
```

```python
loan_fam = pd.read_csv("./대출금리/예금은행 대출금리_가계.csv")
loan_fam = pd.DataFrame(loan_fam)
loan_fam = loan_fam.drop(columns = ["통계표", "단위", "변환"], axis=1)
loan_fam = loan_fam.set_index("계정항목")
loan_fam =loan_fam.T
loan_fam.head(1)
```

```python
base_rate = pd.read_csv("./대출금리/한국은행 기준금리.csv")
base_rate = pd.DataFrame(base_rate)
base_rate = base_rate.drop(columns = ["통계표", "단위", "변환"], axis=1)
base_rate = base_rate.set_index("계정항목")
base_rate =base_rate.T
base_rate
```

```python
loan_com["기준금리"] = base_rate["한국은행 기준금리"]
loan_com.head(1)
```

```python
loan_fam["기준금리"] = base_rate["한국은행 기준금리"]
loan_fam.head(1)
```
