# 예금 & 대출

```python
rate=pd.read_csv("./예금대출.csv")
```

<img src = "./예금대출.png">

```python
예금이자 : 은행이 예금 고객에게 지급하는 이자 (기준금리 + 가산금리 + 우대이율) 
대출이자 : 은행이 대출 고객에게 요구하는 이자 (기준금리 + 가산금리 - 우대이율)

[고객의 신용에 따라 추가 지급 또는 적용되는 금리]
가산금리(Spread Rate) : 고객의 신용에 따라 금융기관 수수료를 추가해 결정
우대금리(Prime Rate) : 신용도가 높은 고객에게 대출할 때 적용하는 금리

# 무위험이자률 (MBS,단기채권)기준 + 0.01 a = 2.52 -> 예대마진 
```

```python
rate.head()
```

```python
rate['차이']=rate.대출이율- rate.예금이율
```

```python
rate.head()
```

```python
rate.차이.mean() #예대마진
```

```python
https://terms.naver.com/entry.naver?docId=1199689&cid=40942&categoryId=31826

예대마진 : 금융기관의 수익성 지표 
금융기관의 수입 (대출이자 - 예금에 지불한 이자) 
```

```python
rate=rate.drop("차이",axis=1)
```

```python
rate.plot(figsize=(30,10))

plt.title('예금 대출 금리 차이 - 예대마진', fontsize= 30)   
plt.xlabel('날짜', fontsize=25)  
plt.ylabel('금리', fontsize=25)
plt.legend(['예금 금리', '대출 금리'], fontsize=25) 


plt.show()
```
