```python
import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings(action='ignore')

import plotly.express as px

pd.set_option('display.max_columns', None)

sns.set()
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family'] = 'Malgun Gothic'
```

# 기준금리

- ecos.bok.or.kr
- 1.3.1. 한국은행 기준금리 및 여수신금리



```python
기준금리_df = pd.read_csv('./한국은행_기준금리_9905_2304.csv', parse_dates=['변환'])
기준금리_df['변환'] = 기준금리_df['변환'].dt.strftime("%Y-%m")
기준금리_df = 기준금리_df.set_index('변환')
# 기준금리_df.index.name=''
기준금리_df.columns = ['기준금리']
# px.line(기준금리_df, y="기준금리")
# fig.show()
기준금리_df.plot(figsize=(14,4))
plt.show()
```

# 원달러환율

- ecos.bok.or.kr
- 3.1.1.1. 주요국 통화의 대원화환율  

<!-- ## 일단위 -->

```python
원달러환율_df = pd.read_csv('./환율_990401.csv', parse_dates=['변환'], thousands=',')
원달러환율_df = 원달러환율_df.set_index('변환')
원달러환율_df.columns = ['원달러환율']
원달러환율_df.plot(figsize=(14,4))
plt.show()
```

```python
원달러환율_df.index[0], 원달러환율_df.index[-1]
```

## 월단위

```python
원달러환율_df_copy = 원달러환율_df.reset_index().copy()
원달러환율_df_copy['변환'] = 원달러환율_df_copy['변환'].dt.strftime("%Y-%m")
원달러환율_df = 원달러환율_df_copy.groupby('변환').mean()
원달러환율_df.plot(figsize=(14,4))
plt.plot()
plt.show()
```

# 환율과 기준금리

```python
df = pd.concat([기준금리_df.loc['1999-05':'2023-04'], 원달러환율_df['1999-05':'2023-04']], axis=1)
df.index = df.index.astype('datetime64')

fig, ax1 = plt.subplots()
fig.set_size_inches(14,4)
ax1.plot(df['기준금리'], color='r')
ax1.legend(['기준금리'])

ax2 = ax1.twinx()
ax2.plot(df['원달러환율'], color='k')
ax2.legend(['원달러환율'])

plt.grid(False)
plt.show()
```

```python
#  2022년 2/4분기 무역수지 적자 기록 
# 상대적으로 달러의 가격 상승 이유 = 달러가 안전자산이라  달러의 수요가 증가했기 떄문 
# 펜데믹 이후 미국이 자이언트 스텝 시행 , 2022년 2/4분기에 주요 원자재 가격이 급상승
```

# 외환보유액

- ecos.bok.or.kr
- 3.5. 외환보유액


```python
외환보유액_df = pd.read_csv('./외환보유액_04100128.csv', parse_dates=['변환'], thousands=',')
외환보유액_df['변환'] = 외환보유액_df['변환'].dt.strftime("%Y-%m")
외환보유액_df = 외환보유액_df.set_index('변환') 
외환보유액_df.columns = ['외환보유액']
외환보유액_df.plot(figsize=(14,4))
plt.show() 
# 외환보유고는 기준금리와 환율과 관계없이 우상
# IMF 구제 금융 역사로 인한 외환보유액 증가
# 펜데믹 이후, 국제 금융 시장 불안 또는 리스크에 선제적으로 대응하기 위해 외환 보유고 증가
```

```python
외환보유액_df.index[0], 외환보유액_df.index[-1]
```

# 기준금리, 원달러환율, 외환보유액

https://www.bok.or.kr/portal/main/contents.do?menuNo=200411 - 한국은행

```python
df = pd.concat([기준금리_df.loc['1999-05':'2023-04'], 원달러환율_df['1999-05':'2023-04'], 외환보유액_df['1999-05':'2023-04']], axis=1)
df.index = df.index.astype('datetime64')

for col in df.columns:
    df[col] = (df[col] - df[col].mean()) / df[col].std()
df.plot(figsize=(14,4))
plt.show()
```

```python
# 감소한 지점의 이유는? - 2008 글로벌 금융위기가 리먼 브라더스 파산하면서 신용 경색,신용리스크를 방지하기 위해 외환보유고 사용 
# 08년도에 환율이 급격히 상승한 이유 = 달러 유동성이 부족해서 달러 가치 상승, 한국은행 및 정부가 한국 시장에 달러 공급해서 외환 보유고가 일시적으로 감소했다.
```

채권가격 = 채권수익률(-)
기준금리 인하 -> 통화량 증가
초과화폐 채권매입
채권 수요 증가 -->채권 가격 상승 --> 채권 수익률 하락 ( 더 높은 가격에 매입해서)

즉, 채권가격의 상승 = 채권수익률(채권금리)하락



```python
# 채권만 있다고 가정
# 처음 현재보다 기준금리 인하 -> 시중통화량 증가 -> 화폐 초과 공급 --> 채권 수요 증가 --> 채권 가격 증가 --> 
# 결론 금리 인하시 수익률 인하 , 채권 가격과 채권수익률은 (-)의 관계
```
