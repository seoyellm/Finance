```python
from scipy import stats
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# 잔존만기 계산

```python
def time_to_maturity(t0, T, workday=252) : 
    # t = pd.to_datetime(시작일).date()
    # T = pd.to_datetime(종료일).date()
    # dcnt = (np.busday_count(T, t))  
    maturity = (np.busday_count(t0, T))/workday
    return maturity
```

```python
time_to_maturity('2018-08-01', '2018-12-14')   
```

# 콜옵션 가격

```python
def call_prices(S,K,ttm,r,sigma):   # ttm = 잔존만기
    d1 = ( np.log(S/K) + (r + sigma**2*0.5)*ttm ) / (sigma * np.sqrt(ttm))
    d2 = ( np.log(S/K) + (r - sigma**2*0.5)*ttm ) / (sigma * np.sqrt(ttm))
    val = ( S * stats.norm.cdf(d1, 0.0, 1.0) ) - K * np.exp(-r*ttm) * stats.norm.cdf(d2,0.0,1.0)
    return val
```

```python
ttm = time_to_maturity('2018-09-01', '2018-12-14')  
r = 0.025
sigma = 0.15
```

```python
call_prices(350,350, ttm,r, sigma)
```

# 기초자산 및 행사가격 변화에 따른 콜옵션가격 변화

```python
call_space = dict()
S = np.arange(200,400,10)
for s in S:
    call_space[s]  = call_prices(s,350, ttm,r, sigma)
# call_space
```

```python
call_space = dict()
K = np.arange(250,350,10)
S = np.arange(200,400,10)
for k in K:
    single_space = dict()
    for s in S:
        single_space[s] = call_prices(s,k,ttm,r,sigma)
    call_space[k] = pd.Series(single_space)
# call_space
```

```python
df_call_space = pd.DataFrame(call_space)
df_call_space[:2]
```

# 풋옵션 가격곡면

```python
def put_prices(S,K,ttm,r,sigma):   # ttm = 잔존만기
    d1 = ( np.log(S/K) + (r + sigma**2*0.5)*ttm ) / (sigma * np.sqrt(ttm))
    d2 = ( np.log(S/K) + (r - sigma**2*0.5)*ttm ) / (sigma * np.sqrt(ttm))
    val = ( K * np.exp(-r * ttm) ) *stats.norm.cdf(-d2,0.0,1.0) - (S * stats.norm.cdf(-d1,0.0,1.0)) 
    return val
```

```python
K = np.arange(250,350,5)
S = np.arange(200,400,10)
K,S = np.meshgrid(K,S)   # 매트릭스 만들기
Z = put_prices(S,K,ttm,r,sigma)

fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')

surf = ax.plot_surface(S,K,Z,cmap = cm.summer
               ,linewidth = 1, antialiased = True, alpha=0.8)

ax.set_xlabel('기초자산')
ax.set_ylabel('행사가격')
ax.set_zlabel('풋옵션가격')

ax.set_title('풋옵션 가격 곡면')

fig.colorbar(surf, shrink = 0.5, aspect=5)
```

# 콜옵션 그릭스

## 콜옵션 델타

```python
def call_delta(S,K,ttm,r,sigma):
    d1 = ( np.log(S/K) + (r + sigma**2*0.5)*ttm ) / (sigma * np.sqrt(ttm))
    val = stats.norm.cdf(d1,0.0,1.0)
    return val
```

```python
K = np.arange(300,350,5)
S = np.arange(200,400,10)
K,S = np.meshgrid(K,S)   # 매트릭스 만들기
Z = call_delta(S,K,ttm,r,sigma)

fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')

surf = ax.plot_surface(S,K,Z,cmap = cm.coolwarm
               ,linewidth = 1, antialiased = True, alpha=0.8)

ax.set_xlabel('기초자산')
ax.set_ylabel('행사가격')
ax.set_zlabel('델타')

ax.set_title('델타 곡면')

fig.colorbar(surf, shrink = 0.5, aspect=5)
```

## 콜옵션 감마

```python
def ndx(x):
    return (  np.exp(-1 * x**2 * 0.5  )  / np.sqrt(2* np.pi))
```

```python
def gamma(S,K,ttm,r,sigma):
    d1 = ( np.log(S/K) + (r + sigma**2*0.5)*ttm ) / (sigma * np.sqrt(ttm))
    val = ( ndx(d1) )  /  (S * sigma * np.sqrt(ttm))
    return val
```

```python
K = np.arange(300,350,5)
S = np.arange(200,400,10)
K,S = np.meshgrid(K,S)   # 매트릭스 만들기
Z = gamma(S,K,0.38,0.02,0.30)

fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')

surf = ax.plot_surface(S,K,Z,cmap = cm.coolwarm
               ,linewidth = 1, antialiased = True, alpha=0.8)

ax.set_xlabel('기초자산')
ax.set_ylabel('행사가격')
ax.set_zlabel('감마')

ax.set_title('감마 곡면')

fig.colorbar(surf, shrink = 0.5, aspect=5)
```

## 콜옵션 세타

```python
def call_theta(S,K,ttm,r,sigma):
    d1 = ( np.log(S/K) + (r + sigma**2*0.5)*ttm ) / (sigma * np.sqrt(ttm))
    d2 = ( np.log(S/K) + (r - sigma**2*0.5)*ttm ) / (sigma * np.sqrt(ttm))
    val = -1 *( (S * ndx(d1)*sigma) / (2*np.sqrt(ttm)) ) - r*K*np.exp(-r*ttm) * stats.norm.cdf(d2,0.0,1.0)
    return val
```

```python

T = np.arange(1.0,0.0,-0.01)
S = np.arange(350,355)
T,S = np.meshgrid(T,S)   # 매트릭스 만들기
Z = call_theta(S,350,T,0.02,0.30)

fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')

surf = ax.plot_surface(T,S,Z,cmap = cm.coolwarm
               ,linewidth = 1, antialiased = True, alpha=0.8)

ax.set_xlabel('잔존만기')
ax.set_ylabel('기초자산')
ax.set_zlabel('쎄타')
ax.invert_xaxis()    # x축을 1부터 0의 순서로 / 거꾸로 출력

ax.set_title('쎄타 곡면')

fig.colorbar(surf, shrink = 0.5, aspect=5)
```

## 콜옵션 베가

```python
def vaga(S,K,ttm, r,sigma):
    d1 = ( np.log(S/K) + (r + sigma**2*0.5)*ttm ) / (sigma * np.sqrt(ttm))
    d2 = ( np.log(S/K) + (r - sigma**2*0.5)*ttm ) / (sigma * np.sqrt(ttm))
    val =( S * np.sqrt(ttm) * ndx(d1) )
    return val
    
```

```python

V = np.arange(0.05,0.95,0.01)
S = np.arange(350,352)
V,S = np.meshgrid(V,S)   # 매트릭스 만들기
Z = vaga(S,350,0.38,0.02,V)

fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')

surf = ax.plot_surface(V,S,Z,cmap = cm.coolwarm
               ,linewidth = 1, antialiased = True, alpha=0.8)

ax.set_xlabel('잔존만기')
ax.set_ylabel('기초자산')
ax.set_zlabel('베타')
# ax.invert_xaxis()    # x축을 1부터 0의 순서로 / 거꾸로 출력

ax.set_title('베타 곡면')

fig.colorbar(surf, shrink = 0.5, aspect=5)
```
