### 정적 비중 투자, 분기별 리밸런싱

* WeighSpecified(..) : 정적 자산배분
* <font color=red><b>RunQuarterly() : 분기별 리밸런싱을 통해 투자비중 일정하게 유지

```python
data = prices[['SPY', 'TLT', 'IEF','GLD','DBC']].dropna()
aw = bt.Strategy('All Weather',
                 [
                    bt.algos.SelectAll(),
                    bt.algos.WeighSpecified(SPY=0.3, TLT=0.4, IEF=0.15, GLD=0.075, DBC=0.075),
                    bt.algos.RunQuarterly(),
                    bt.algos.Rebalance() 
                 ]
                )
aw_backtest = bt.Backtest(aw, data)
aw_result   = bt.run(aw_backtest)

aw_result.plot(figsize=(10,3), title='All Weather', legend=False)
plt.show()
```

## 모멘텀 전략 예
* ref : https://github.com/pmorissette/bt/issues/189
* 투자 종목 중 과거 12개월 수익률이 높은 5개 자산을 선택
* 위험균형 포트폴리오를 구성
* 매월 말 리밸런싱

* SelectMomentum() : 과거 years=n 동안 수익률 Top-N
* WeighERC() : 위험균형 포트폴리오를 구성
* RunMonthly() : 매월말 리밸런싱
* Rebalance() : 계산된 비중에 따라 리밸런싱    

```python
gdaa = bt.Strategy('Momentum', [
    bt.algos.SelectAll(),
    bt.algos.SelectMomentum(n=5, lookback=pd.DateOffset(years=1)),
    bt.algos.WeighERC(lookback=pd.DateOffset(years=1)),
    bt.algos.RunMonthly(),    
    bt.algos.Rebalance()
])

gdaa_backtest = bt.Backtest(gdaa, data)
gdaa_result = bt.run(gdaa_backtest)
my_stats_result(gdaa_result)
gdaa_result.plot(figsize=(10, 3),
                 title='Global Dynamic Asset Allocation',
                 legend=False)
plt.show()
```

## 추세추종 이동평균선 전략
* 추세추종 전략을 사용할 경우 하락 방어가 우수
* 매수(GC) : 주가 > 이동평균
* 매도(DC) : 주가 < 이동평균

```python
import talib
# SMA
data = prices[['SPY']].dropna()
print(data.columns)
sma = data.apply(lambda x: talib.SMA(x, 200))
print(sma[-5:])
```

* SelectWhere(data > sma) : 조건 식 만족 시 매수

```python
bt_sma = bt.Strategy('bt_sma', [
                    bt.algos.SelectWhere(data > sma),
                    bt.algos.WeighEqually(),
                    bt.algos.Rebalance()
                ])

bt_sma_backtest = bt.Backtest(bt_sma, data)
```

```python
def timing(price, n):
    sma = price.apply(lambda x: talib.SMA(x, n))
    stragety = bt.Strategy(n, [
        bt.algos.SelectWhere(price > sma),        
        bt.algos.WeighEqually(),
        bt.algos.Rebalance()
    ])

    backtest = bt.Backtest(stragety, price)
    return (backtest)

n_20  = timing(data, 20)
n_60  = timing(data, 60)
n_100 = timing(data, 100)
n_150 = timing(data, 150)
n_200 = timing(data, 200)
n_250 = timing(data, 250)

result = bt.run(n_20, n_60, n_100, n_150, n_200, n_250)
my_stats_result(result)
result.prices[250:].rebase().plot(figsize=(10, 3))
plt.show()
```

## 롱숏전략 : 양방향 매매
* 롱(골든크로스)  : 단기이평 > 장기이평 
* 숏(데드크로스)  : 단기이평 < 장기이평  : 공매도, 인버스, 매도 포지션

```python
data = prices[['SPY']]
SMA_200 = data.apply(lambda x: talib.SMA(x, 200))
SMA_60 = data.apply(lambda x: talib.SMA(x, 60))
signal = SMA_200.copy()
signal[SMA_60 >= SMA_200] = 1
signal[SMA_60 < SMA_200] = -1
signal[signal.isnull()] = 0
```

```python
bind = pd.concat([data, SMA_200, SMA_60, signal], axis=1)
bind.columns = ['SPY', 'SMA 200', 'SMA 60', 'signal']
bind.loc['2018':].plot(figsize=(10, 3), secondary_y=['signal'])
plt.show()
```

```python
strategy = bt.Strategy(
    'SMA_crossover',
    [bt.algos.SelectAll(),
     bt.algos.WeighTarget(signal),
     bt.algos.Rebalance()])
backtest = bt.Backtest(strategy, data)
result = bt.run(backtest)
result.plot(figsize=(10, 3))
plt.show()
```

# 평균회귀 전략

## RSI를 이용한 전략
* RSI : 매수 매도의 상대적 강도를 나타낸 지표
* RSI >= 70 (과매수) : 매도
* RSI <= 30 (과매도) : 매수

```python
data = prices[['SPY']]
spy_rsi = data.apply(lambda x: talib.RSI(x, 14))

# -1(숏 포지션),  1(롱 포지션),  0(30~70)
signal = spy_rsi.copy()
signal[spy_rsi > 70] = -1
signal[spy_rsi < 30] = 1
signal[(spy_rsi <= 70) & (spy_rsi >= 30)] = 0
signal[signal.isnull()] = 0
```

```python
bind = pd.concat([data, signal], axis=1)
bind.columns = ['SPY', 'signal']
bind.loc['2018':].plot(figsize=(10, 3), secondary_y=['signal'])
plt.show()
```

```python
from matplotlib import gridspec

fig = plt.subplots(figsize=(10, 3), sharex=True)
gs = gridspec.GridSpec(nrows=2, ncols=1, height_ratios=[2, 1])

ax1 = plt.subplot(gs[0])
ax1 = data['SPY'].plot()
ax1.set_xlabel('')
ax1.axes.xaxis.set_ticks([])

ax2 = plt.subplot(gs[1])
ax2 = spy_rsi['SPY'].plot(color='black', ylim=[0, 100])
ax2 = plt.axhline(y=30, color='red', linestyle='-')
ax2 = plt.axhline(y=70, color='red', linestyle='-')

plt.subplots_adjust(wspace=0, hspace=0)
plt.show()
```

```python
strategy = bt.Strategy('RSI backtesting',
                       [bt.algos.WeighTarget(signal),
                        bt.algos.Rebalance()])
backtest = bt.Backtest(strategy, data)
result = bt.run(backtest)
my_stats_result(result)
result.plot(figsize=(10, 3))
plt.show()
```

```python
pd.concat([bind, result.prices],
          axis=1).loc['2020'].rebase().plot(figsize=(10, 3),
                                            secondary_y=['signal'],
                                            alpha=0.8)
plt.show()
```

## 볼린저밴드 전략
*  이동평균선을 중심으로 일정 표준편차를 상한선과 하한선으로 설정한 밴드
* 주가 > 상한밴드 (과매수 상태) : 매도
* 주가 < 하한밴드 (과매도 상태) : 매수

```python
data = prices[['SPY']]
upper_2sd, mid_2sd, lower_2sd = talib.BBANDS(data['SPY'],
                                             nbdevup=2,
                                             nbdevdn=2,
                                             timeperiod=20)

bb = pd.concat([upper_2sd, mid_2sd, lower_2sd, data['SPY']], axis=1)
bb.columns = ['Upper Band', 'Mid Band', 'Lower Band', 'SPY']
```

```python
signal.shape, bb.shape
```

```python
# bb.head(1), signal.head(1)
```

```python
signal = data.copy()
signal['SPY'] = 0
signal[bb['SPY'] > bb['Upper Band']] = -1
signal[bb['SPY'] < bb['Lower Band']] = 1
signal[signal.isnull()] = 0
```

```python
strategy = bt.Strategy('BB',
                       [bt.algos.WeighTarget(signal),
                        bt.algos.Rebalance()])
backtest = bt.Backtest(strategy, data)
result = bt.run(backtest)
my_stats_result(result)
result.plot(figsize=(10, 3))
plt.show()
```

# 바이 앤 홀드 전략

* bt.algos.RunOnce() : 리밸런싱 없이 처음 상태가 그대로 유지

```python
def buy_and_hold(data, name):
    bt_strategy = bt.Strategy(name, [        
        bt.algos.SelectAll(),
        bt.algos.WeighEqually(),
        bt.algos.RunOnce(),
        bt.algos.Rebalance()
    ])
    return bt.Backtest(bt_strategy, data)

# 벤치마크 전략 백테스트
buy_hold_backtest = buy_and_hold(data[['SPY']], name='buy_hold')

# 두개 백테스트 동시에 실행
bt_sma_result = bt.run(bt_sma_backtest, buy_hold_backtest)
my_stats_result(bt_sma_result)
```
